```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#https://blog.softhints.com/pandas-display-all-columns-and-show-more-rows/
import numpy as np
df_surveymonkey = pd.read_csv ('survey-monkey2.csv',skiprows=0)
#df_finaldata=pd.read_csv ('final_data.csv')
```

## Does the organisation Experience staff turnover?
### Number of Employees likely to Exit the organisation

```python
df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']
attrition_is_yes= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "No"]
attrition_is_no= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "Yes"]
```

```python
#define data
#https://www.statology.org/seaborn-pie-chart/
yes=len(attrition_is_yes)
no=len(attrition_is_no)
data = [yes,no]
labels = ['Attrition Yes', 'Attrition No']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

### From the analysis, 20% of the total sample are likely to exit the organisation in the next 3 years

```python
attrition_is_yes = attrition_is_yes.rename(columns={'I see myself working at the Agriculture and Food Authority in the next 2 years ?': 'Attrition','The organization has a tool for monitoring and evaluating employee performance?':'performance-monitoring','The organization is supportive in ensuring you achieve your strategic objectives?':'performance-support','The organization has a strategy of rewarding good performance and improving poor performance?':'performance-reward','I was able to achieve my departmental targets for the last one year?':'target-achieved'})
attrition_is_yes = attrition_is_yes[['Attrition','performance-monitoring','performance-support','performance-reward','target-achieved']]
```

```python
training_offered_yes=attrition_is_yes[attrition_is_yes['performance-monitoring']== "Yes"]
training_offered_NO=attrition_is_yes[attrition_is_yes['performance-monitoring']== "No"]
yes=len(training_offered_yes)
no=len(training_offered_NO)
data = [yes,no]
labels = ['performance monitoring', 'No performance monitoring ']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
performance_support_yes=attrition_is_yes[attrition_is_yes['performance-support']== "Yes"]

performance_support=attrition_is_yes[attrition_is_yes['performance-support']== "No"]

yes=len(performance_support_yes)
no=len(performance_support)
data = [yes,no]
labels = ['Performance support', 'No performance support']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()

```

```python
performance_reward=attrition_is_yes[attrition_is_yes['performance-reward']== "Yes"]

performance_reward_no=attrition_is_yes[attrition_is_yes['performance-reward']== "No"]

yes=len(performance_reward)
no=len(performance_reward_no)
data = [yes,no]
labels = ['Performance Reward Offered', 'No Performance Reward']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
target_achieved=attrition_is_yes[attrition_is_yes['target-achieved']== "Yes"]

target_achieved_no=attrition_is_yes[attrition_is_yes['target-achieved']== "No"]

yes=len(target_achieved)
no=len(target_achieved_no)
data = [yes,no]
labels = ['Target Achieved', 'Target not Achieved']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
