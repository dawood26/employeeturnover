```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#https://blog.softhints.com/pandas-display-all-columns-and-show-more-rows/
import numpy as np
df_surveymonkey = pd.read_csv ('survey-monkey2.csv',skiprows=0)
#df_finaldata=pd.read_csv ('final_data.csv')
```

## Does the organisation Experience staff turnover?
### Number of Employees likely to Exit the organisation

```python
df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']
attrition_is_yes= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "No"]
attrition_is_no= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "Yes"]
```

```python
#define data
#https://www.statology.org/seaborn-pie-chart/
yes=len(attrition_is_yes)
no=len(attrition_is_no)
data = [yes,no]
labels = ['Attrition Yes', 'Attrition No']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

### From the analysis, 20% of the total sample are likely to exit the organisation in the next 3 years

```python
attrition_is_yes = attrition_is_yes.rename(columns={'I see myself working at the Agriculture and Food Authority in the next 2 years ?': 'Attrition','I understand my career path':'careerpath','My job placement is in line with my career training':'job_placement_relevant','My skills are fully recognized and utilized':'skills_utilized'})
attrition_is_yes = attrition_is_yes[['Attrition','careerpath','job_placement_relevant','skills_utilized']]
```

```python
clear_career_path=attrition_is_yes[attrition_is_yes['careerpath']== "Yes"]
clear_career_path_no=attrition_is_yes[attrition_is_yes['careerpath']== "No"]
yes=len(clear_career_path)
no=len(clear_career_path_no)
data = [yes,no]
labels = ['Clear career path', 'No clear career path']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
job_placement_relevant=attrition_is_yes[attrition_is_yes['job_placement_relevant']== "Yes"]

job_placement_relevant_no=attrition_is_yes[attrition_is_yes['job_placement_relevant']== "No"]

yes=len(job_placement_relevant)
no=len(job_placement_relevant_no)
data = [yes,no]
labels = ['Job is relevant', 'Job placement not relevant']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()

```

```python
skills_utilized=attrition_is_yes[attrition_is_yes['skills_utilized']== "Yes"]

skills_utilized_no=attrition_is_yes[attrition_is_yes['skills_utilized']== "No"]

yes=len(skills_utilized)
no=len(skills_utilized_no)
data = [yes,no]
labels = ['My skills are utilized', 'My Skills not Utilized']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
