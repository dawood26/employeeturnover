```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#https://blog.softhints.com/pandas-display-all-columns-and-show-more-rows/
import numpy as np
df_surveymonkey = pd.read_csv ('survey-monkey2.csv',skiprows=0)
#df_finaldata=pd.read_csv ('final_data.csv')
```

## Does the organisation Experience staff turnover?
### Number of Employees likely to Exit the organisation

```python
df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']
attrition_is_yes= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "No"]
attrition_is_no= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "Yes"]
```

```python
#define data
#https://www.statology.org/seaborn-pie-chart/
yes=len(attrition_is_yes)
no=len(attrition_is_no)
data = [yes,no]
labels = ['Attrition Yes', 'Attrition No']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

### From the analysis, 20% of the total sample are likely to exit the organisation in the next 3 years

```python
attrition_is_yes = attrition_is_yes.rename(columns={'I see myself working at the Agriculture and Food Authority in the next 2 years ?': 'Attrition','I have enough data and information to support decision making within the department and it is readily available for inference?':'enough_data_decision','The department has a system that allows for high level analysis and visualization of data to support decision-making?':'data_analysis_system','Long term decisions made within your department are based on factual analyzed data?':'information_based_decisions'})
attrition_is_yes = attrition_is_yes[['Attrition','enough_data_decision','data_analysis_system','information_based_decisions']]
```

```python
enough_data_decision=attrition_is_yes[attrition_is_yes['enough_data_decision']== "Yes"]
enough_data_decision_no=attrition_is_yes[attrition_is_yes['enough_data_decision']== "No"]
yes=len(enough_data_decision)
no=len(enough_data_decision_no)
data = [yes,no]
labels = ['Enough data available for decision making', 'Data not enough for decision making']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
data_analysis_system=attrition_is_yes[attrition_is_yes['data_analysis_system']== "Yes"]

data_analysis_system_no=attrition_is_yes[attrition_is_yes['data_analysis_system']== "No"]

yes=len(data_analysis_system)
no=len(data_analysis_system_no)
data = [yes,no]
labels = ['High level Data Analysis', 'No High level Data Analysis']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()

```

```python
information_based_decisions=attrition_is_yes[attrition_is_yes['information_based_decisions']== "Yes"]

information_based_decisions_no=attrition_is_yes[attrition_is_yes['information_based_decisions']== "No"]

yes=len(information_based_decisions)
no=len(information_based_decisions_no)
data = [yes,no]
labels = ['Decisions based on Factual Data', 'Decisions not based on Factual Data']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
