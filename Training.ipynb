```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#https://blog.softhints.com/pandas-display-all-columns-and-show-more-rows/
import numpy as np
df_surveymonkey = pd.read_csv ('survey-monkey2.csv',skiprows=0)
#df_finaldata=pd.read_csv ('final_data.csv')
```

## Does the organisation Experience staff turnover?
### Number of Employees likely to Exit the organisation

```python
df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']
attrition_is_yes= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "No"]
attrition_is_no= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "Yes"]
```

```python
#define data
#https://www.statology.org/seaborn-pie-chart/
yes=len(attrition_is_yes)
no=len(attrition_is_no)
data = [yes,no]
labels = ['Attrition Yes', 'Attrition No']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

### From the analysis, 20% of the total sample are likely to exit the organisation in the next 3 years

```python
attrition_is_yes = attrition_is_yes.rename(columns={'I see myself working at the Agriculture and Food Authority in the next 2 years ?': 'Attrition','The organization is doing enough to ensure the employees undergo training and other professional development activities?':'Trainings-offered','The training opportunities offered by the organization are based on skills gap and improving performance targets?':'Trainings-relevant','I have attended one or more trainings which have led to a promotion ?':'Promotion-after-training'})
attrition_is_yes = attrition_is_yes[['Attrition','Trainings-offered','Trainings-relevant','Promotion-after-training']]
```

### a) 46% of the employees likely to exit the organisation,say the organisation did not offer them training training opportunities 

```python
training_offered_yes=attrition_is_yes[attrition_is_yes['Trainings-offered']== "Yes"]
training_offered_NO=attrition_is_yes[attrition_is_yes['Trainings-offered']== "No"]
yes=len(training_offered_yes)
no=len(training_offered_NO)
data = [yes,no]
labels = ['Training Offered', 'Training Not offered']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

### b) 77% of the employees likely to exit the organisation,say the organisation did not offer them relevant training opportunities 

```python
training_offered_relevant=attrition_is_yes[attrition_is_yes['Trainings-relevant']== "Yes"]

training_offered_irrelevant=attrition_is_yes[attrition_is_yes['Trainings-relevant']== "No"]

yes=len(training_offered_relevant)
no=len(training_offered_irrelevant)
data = [yes,no]
labels = ['Trainings relevant', 'Trainings not relevant']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()

```

```python
promotion_offered=attrition_is_yes[attrition_is_yes['Promotion-after-training']== "Yes"]

promotion_not_offered=attrition_is_yes[attrition_is_yes['Promotion-after-training']== "No"]

yes=len(promotion_offered)
no=len(promotion_not_offered)
data = [yes,no]
labels = ['Promotion Offered', 'Promotion not offer']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
