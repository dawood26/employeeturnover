```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#https://blog.softhints.com/pandas-display-all-columns-and-show-more-rows/
import numpy as np
df_surveymonkey = pd.read_csv ('survey-monkey2.csv',skiprows=0)
#df_finaldata=pd.read_csv ('final_data.csv')
```

## Does the organisation Experience staff turnover?
### Number of Employees likely to Exit the organisation

```python
df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']
attrition_is_yes= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "No"]
attrition_is_no= df_surveymonkey[df_surveymonkey['I see myself working at the Agriculture and Food Authority in the next 2 years ?']== "Yes"]
```

```python
#define data
#https://www.statology.org/seaborn-pie-chart/
yes=len(attrition_is_yes)
no=len(attrition_is_no)
data = [yes,no]
labels = ['Attrition Yes', 'Attrition No']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

### From the analysis, 20% of the total sample are likely to exit the organisation in the next 3 years

```python
attrition_is_yes = attrition_is_yes.rename(columns={'I see myself working at the Agriculture and Food Authority in the next 2 years ?': 'Attrition','The organization has a system for vacancy identification and recruitment process?':'clear_recruitment_process','During my job interview, I was able to communicate information about my relevant skills, experience, and career goals?':'communicate_properly_interview','In overall, I was satisfied with the recruiting process for this position?':'satisfied_with_recruitment'})
attrition_is_yes = attrition_is_yes[['Attrition','clear_recruitment_process','communicate_properly_interview','satisfied_with_recruitment']]
```

```python
clear_recruitment_process=attrition_is_yes[attrition_is_yes['clear_recruitment_process']== "Yes"]
clear_recruitment_process_no=attrition_is_yes[attrition_is_yes['clear_recruitment_process']== "No"]
yes=len(clear_recruitment_process)
no=len(clear_recruitment_process_no)
data = [yes,no]
labels = ['Clear recruitment process', 'No clear recruitment process']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
communicate_properly_interview=attrition_is_yes[attrition_is_yes['communicate_properly_interview']== "Yes"]

communicate_properly_interview_no=attrition_is_yes[attrition_is_yes['communicate_properly_interview']== "No"]

yes=len(communicate_properly_interview)
no=len(communicate_properly_interview_no)
data = [yes,no]
labels = ['Communicated properly During interview', 'No proper chance to communicate']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()

```

```python
satisfied_with_recruitment=attrition_is_yes[attrition_is_yes['satisfied_with_recruitment']== "Yes"]

satisfied_with_recruitment_no=attrition_is_yes[attrition_is_yes['satisfied_with_recruitment']== "No"]

yes=len(satisfied_with_recruitment)
no=len(satisfied_with_recruitment_no)
data = [yes,no]
labels = ['Satisfied with recruitment process', 'Not satisfied with recruitment process']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
