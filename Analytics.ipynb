```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#https://blog.softhints.com/pandas-display-all-columns-and-show-more-rows/
import numpy as np
df = pd.read_csv ('final_data.csv')
#df
```

## Does the organisation Experience staff turnover?
### Number of Employees likely to Exit the organisation
 


```python
attrition_is_yes= df[df['C-6e-IV']== "NO"]
attrition_is_no= df[df['C-6e-IV']== "YES"]
#attrition_is_yes
```

```python
#define data
#https://www.statology.org/seaborn-pie-chart/
yes=len(attrition_is_yes)
no=len(attrition_is_no)
data = [yes,no]
labels = ['Attrition Yes', 'Attrition No']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

### From the analysis, 29% of the total sample are likely to exit the organisation in the next 3 years

## 1 . Employee Training and Development Strategy

```python
attrition_is_yes = attrition_is_yes.rename(columns={'C-6e-IV': 'Attrition','C-6a-I':'Trainings-offered','C-6a-II':'Trainings-relevant','C-6a-IV':'Promotion-after-training'})
attrition_is_yes = attrition_is_yes[['Attrition','Trainings-offered','Trainings-relevant','Promotion-after-training']]
```

### 

### a) 83% of the employees likely to exit the organisation,say the organisation did not offer them training training opportunities 

```python
training_offered_yes=attrition_is_yes[attrition_is_yes['Trainings-offered']== "YES"]
training_offered_NO=attrition_is_yes[attrition_is_yes['Trainings-offered']== "NO"]
yes=len(training_offered_yes)
no=len(training_offered_NO)
data = [yes,no]
labels = ['Training Offered', 'Training Not offered']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

### b) 75% of the employees likely to exit the organisation,say the organisation did not offer them relevant training opportunities 

```python
training_offered_yes=attrition_is_yes[attrition_is_yes['Trainings-relevant']== "YES"]
training_offered_NO=attrition_is_yes[attrition_is_yes['Trainings-relevant']== "NO"]
yes=len(training_offered_yes)
no=len(training_offered_NO)
data = [yes,no]
labels = ['Trainings-relevant', 'Trainings Not relevant']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```

### b) 92% of the employees likely to exit the organisation were never promoted after undergoing training 

```python
training_offered_yes=attrition_is_yes[attrition_is_yes['Promotion-after-training']== "YES"]
training_offered_NO=attrition_is_yes[attrition_is_yes['Promotion-after-training']== "NO"]
yes=len(training_offered_yes)
no=len(training_offered_NO)
data = [yes,no]
labels = ['Promoted after Trainings', 'Not promoted after trainings']
#define Seaborn color palette to use
colors = sns.color_palette('bright')[0:5]
#create pie chart
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
